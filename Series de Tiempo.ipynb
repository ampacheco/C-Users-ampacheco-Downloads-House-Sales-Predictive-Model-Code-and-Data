```python
import pandas as pd

data = pd.read_csv("./dolar_2019-2020.csv")
data["Date"] = pd.to_datetime(data["Date"])
data.head()
data = data.dropna()

# create 12 month moving average
data["Value12"] = data["Value"].rolling(12).mean()
```

```python
data.tail()
```

```python
data.dtypes
```

```python
# Plot
import plotly.express as px

fig = px.line(data, x='Date', y=['Value', 'Value12'], template = 'plotly_dark')

fig.show()
```
